```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Generate a sequence of 50 values from 0 to 10
x = np.linspace(0, 10)
y = np.sin(x)
z = np.cos(x)

# Construct the data visualization chart
plt.figure(figsize=(8, 4))
plt.plot(x, y, 'b', label='Sin(x)')
plt.plot(x, z, 'r--', label='Cos(x)')

# Add formatting labels
plt.xlabel('Radians')
plt.ylabel('Value')
plt.title('Engineering Signal Analysis')
plt.legend()
plt.grid(True)
plt.show()
```
